```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load the dataset
spending=pd.read_csv("customer_spending.csv")
```

```python
spending.info()
```

```python
#get statistical description of the data
spending.describe()
```

```python
#display the first 5 rows of the dataframe
spending.head()
```

```python
#check whether there are duplicates
spending.duplicated().sum()
```

```python
#rename the column Spending(Range 1-100) to Spending without creating new obj
spending.rename(columns={'Spending (Range 1-100)':'Spending'},inplace=True)
sns.lineplot(x=spending["Age"],y=spending["Spending"])
plt.title("The rlnship between customer spending and Age")
```

```python
plt.plot(spending["Age"],spending["Income in (1000)"])
plt.title("The rlnship between customer income and Age")
plt.ylabel('Age')
plt.xlabel('Income in (1000)')
plt.show()
```

```python
sns.barplot(x=spending["Region"], y=spending["Spending"], 
data=spending["Spending"])
#set the title for the chart
plt.title("Region nad Spending")
plt.show()
```

```python
sns.barplot(x=spending["Gender"], y=spending["Spending"], 
data=spending["Spending"])
#set the title for the chart
plt.title("Gender and Spending")
plt.show()
```

```python
sns.barplot(x=spending["Region"], y=spending["Income in (1000)"], 
data=spending["Income in (1000)"])
#set the title for the chart
plt.title("Region and Income")
plt.show()
```
