```python
#importing libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the data
UKdata =pd.read_csv('UK_Accident.csv')
```

```python
#Displaya few rows of the data
UKdata.head()
```

**Get the datatype for each column**

```python
#Exploring the data
UKdata.info()
```

```python
#statistical view of the data
UKdata.describe()
```

**The dataset is balanced as per the count**

```python
#remove the unnamed column
UKdata.drop(UKdata.columns[UKdata.columns.str.contains('Unnamed',case = False)],axis = 1, inplace = True)
```

```python
#Check whether there are null values
UKdata.isnull().sum()
```

**LSOA_of_Accident_Location column has a big percentage of missing values values**

```python
#Drop a column that have several missing values and all the other missing values
UKdata.drop('LSOA_of_Accident_Location', axis=1, inplace = True)
UKdata.dropna(inplace = True)
```

```python
#check whether there are duplicates
UKdata.duplicated().sum()
```

**There are 34155 duplicates. The next thing is to remove the duplicates**


```python
#remove the duplicates
UKdata.drop_duplicates(inplace = True)
```

```python
#visualize which day of the week did many accidents occur
plt.figure(figsize = (8,6))
plt.title('Number of accidents as per the day of week')
sns.countplot(x='Day_of_Week', data = UKdata,order = UKdata['Day_of_Week'].value_counts().index[:20] ,linewidth =5)
```

**Most accidents occured on day 6**

```python
#Accident severity
plt.hist(UKdata['Accident_Severity'])
plt.title('Accident_Severity')
plt.show() 
```

**Most accidents that occured had high severity**

**The number of accidents have been decreasing from the year 2005 to 2013 but went up again in the year 2014**

```python
#Accident severity based on road type
plt.figure(figsize=(15,7))
ax = sns.countplot(x = 'Road_Type',hue='Accident_Severity',data=UKdata, 
                   order = UKdata["Road_Type"].value_counts().index, palette = "YlGnBu")
ax.set_xticklabels(['Single carriageway','Dual_carriageway','Roundabout',
                    'One_way_street ',
                    'Slip road','Unknown'])
plt.legend(['Fatal','Serious','Slight'])
plt.ylabel("Frequency", fontsize = 14)
plt.title('Accident severity based on road type')
plt.show()
```

**Most accidents occurred in a single carriage way**

# Conclusion

From the above visualization, the following insights were drawn:
    The number of causalities increase with increase in the number of vehicles.
    Most accidents that occured had high severity.
    The number of accidents have been decreasing from the year 2005 to 2013 but went up again in the year 2014.
    Most accidents occurred in a single carriage way.
    Few accidents occured in Rural as compared to urban.
    From the above barplot, most accidents occur when the speed limit is 30km/h.
    Police attended most of the accidents.
    Most accidents occured on day 6.
    Most accidents occur at 1700hrs.
    Most accidents occur under normal conditions, i.e. Road surface is Dry, weather is fine, enough light,no special conditions, no junction controls, no Pedestrian_Crossing-Human_Control within 50 metre, no physical crossing within 50 meters.
