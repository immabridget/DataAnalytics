```python
columns = ['text','author_id', 'Location', 'attachments', 'conversation_id', 'created_at', 
           'entities', 'id', 'in_reply_to_user_id', 'lang', 'public_metrics', 
             'possibly_sensitive', 'referenced_tweets', 'reply_settings', 
           'source', 'withheld']
data = []
for tweet in tweets.data:
    data.append([tweet.text, tweet.author_id, tweet.geo, tweet.attachments, tweet.conversation_id, tweet.created_at,
                 tweet.entities, tweet.id, tweet.in_reply_to_user_id, tweet.lang, tweet.public_metrics, 
                 tweet.possibly_sensitive, tweet.referenced_tweets, tweet.reply_settings, tweet.source,
                 tweet.withheld])
    
df = pd.DataFrame(data, columns = columns)
df.to_csv('Elections1.csv')
```
