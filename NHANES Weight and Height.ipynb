```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load the dataset
df=pd.read_csv("NHANES Weight and Height.csv")
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
#remove the unnamed column
df.drop(df.columns[df.columns.str.contains('Unnamed',case = False)],axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# height, weight scatter plot
plt.figure(figsize = (10,12))
sns.scatterplot(data=df, x = "Weight (kg)", y = "Standing Height (cm)",palette = "winter")
plt.show()
```

```python
df.rename(columns={'Weight (kg)':'Weight'},inplace=True)
df.rename(columns={'Standing Height (cm)':'Height'},inplace=True)
df.rename(columns={'BMI(kg/m**2)':'BMI'},inplace=True)

sns.lineplot(x=df["Weight"],y=df["BMI"])
sns.lineplot(x=df["Height"],y=df["BMI"])
plt.title("The rlnship between Height,Weight and BMI")
```
